```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import pickle
```

```python
info = pickle.load(open('/Users/duke/PycharmProjects/rlplay/deep_rl_project/algos/info_1995', 'rb'))

iter_reward = [info[i][0] for i in range(len(info))]
avg_loss1 = [info[i][1] for i in range(len(info))]
avg_loss2 = [info[i][2] for i in range(len(info))]
avg_obj = [info[i][3] for i in range(len(info))]
```

```python
plot_title = "TD3 Training"
data = [iter_reward, avg_loss1, avg_loss2, avg_obj]
sub_titles = ["Reward", "Q1 Function Loss", "Q2 Function Loss", "Policy Objective"]
x_labels = ["Iterations (200 steps)" for i in range(len(data))]
y_labels = ["Reward", "Q1 Loss", "Q2 Loss", "Objective"]

fig, axs = plt.subplots(1,len(data), figsize=(20,4))
fig.suptitle(plot_title)

axs = axs.flatten()
for i in range(len(data)):
    axs[i].set_title(sub_titles[i])
    axs[i].set_xlabel(x_labels[i])
    axs[i].set_ylabel(y_labels[i])
    axs[i].plot(range(len(data[i])), data[i])
```
